# [dbo].[vw_tickets_full]

```sql
/****** Object:  View [dbo].[vw_tickets_full]    Script Date: 16/03/2026 14:07:32 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE   VIEW [dbo].[vw_tickets_full] AS
SELECT
    t.*,
    tt.name               AS ticket_type_name,
    h.name                AS hotel_name,
    h.code                AS hotel_code,
    uc.name               AS created_by_name,
    ua.name               AS assigned_to_name,
    s.title               AS suggestion_title,
    s.status              AS suggestion_status
FROM dbo.tickets t
LEFT JOIN dbo.ticket_types tt ON t.ticket_type_id = tt.id
LEFT JOIN dbo.hotels h        ON t.hotel_id       = h.id
LEFT JOIN dbo.users uc        ON t.created_by    = uc.id
LEFT JOIN dbo.users ua        ON t.assigned_to   = ua.id
LEFT JOIN dbo.suggestions s  ON t.suggestion_id = s.id;
GO
```

# [dbo].[vw_suggestions_full]

```sql
/****** Object:  View [dbo].[vw_suggestions_full]    Script Date: 16/03/2026 14:07:32 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE VIEW [dbo].[vw_suggestions_full] AS
SELECT 
    s.id,
    s.title,
    s.description,
    s.benefits,
    s.status,
    s.cancellation_reason,
    s.created_at,
    s.updated_at,
    uc.name            AS created_by_name,
    ua.name            AS assigned_to_name,
    p.id               AS project_id,
    p.name             AS project_name,
    p.version          AS project_version,
    p.status           AS project_status
FROM dbo.suggestions s
INNER JOIN dbo.users uc   ON s.created_by   = uc.id
LEFT   JOIN dbo.users ua   ON s.assigned_to  = ua.id
LEFT   JOIN dbo.projects p ON s.project_id   = p.id;
GO
```
